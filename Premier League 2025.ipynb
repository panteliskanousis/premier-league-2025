#### ADD THE COLUMN 'ASSISTS_PER_GAME' TO THE DATAFRAME 'PLAYERS'

```python
players['assists_per_game_ratio']=players['assists']/players['played']
```

### REARRANGING THE COLUMN I ADDED IN THE PREVIOUS LINE (ASSISTS_PER_GAME) 

```python
cols=list(players.columns)
players=players[cols[0:16] + [cols[-1]] + cols[16:20]]
```

```python
players = players.loc[:, ~players.columns.duplicated()] #THE COLUMN assist_per_game ADDED 4 TIMES BY MISTAKE SO THIS IS THE WAY TO DELETE THE 3 DUPLICATED COLUMNS AND KEEP THE 1 
 


```

### ADD THE COLUMN 'GOALS PER GAME' TO THE DATAFRAME 'PLAYERS'

```python
players['goals_per_game']=players['goals']/players['played'].replace(0,1) #COMMAND REPLACE ITS VERY USEFUL WHEN WE DONT KNOW IF THE DENOMINATOR MIGHT BE ZERO 
```

### ADD A COLUMN THAT SUMS THE GOALS AND ASSISTS FOR EACH PLAYERS

```python
players['goals+assists']=players['goals']+players['assists']
```

### REARRANGING THE COLUMN I ADDED IN THE PREVIOUS LINE (ASSISTS_PER_GAME)

```python
cols=list(players.columns)
players=players[cols[0:11] + [cols[-1]] + cols[11:24]]
```

```python
players = players.loc[:, ~players.columns.duplicated()]
```

### THE SUPER SUBS(THE PLAYERS THAT SCORING OFTEN AS SUBSTITUTES

```python
players['minutes/played']=players['minutes']/players['played']
players['goals/minutes']=players['goals']/players['minutes']*90
super_subs=players[(players['minutes/played']<=45) & (players['goals/minutes']>0.5)&(players['goals']>=3) & (players['played']>=8)]
super_subs.nlargest(5,'goals')[['name','goals','played','starts','minutes','goals/minutes','minutes/played']]
```

### SCATTERPLOT FOR SUPER SUBS THAT X=AVERAGE MINUTES PER GAME AND Y= GOALS PER 90 MINUTES

```python
plt.figure(figsize=(8,7))
plt.scatter(players['minutes/played'],players['goals/minutes'],color='lightgray',s=40,ec='black')
plt.scatter(super_subs['minutes/played'],super_subs['goals/minutes'],color='red',s=100,ec='black')
plt.xlabel('Average Minutes per Game')
plt.ylabel('Goals per 90 Minutes')
plt.title('Super Subs')
plt.show()
```

### THE TEAM WITH THE MOST BALANCED OFFENSE (THE TEAM THAT HAS THE MOST PLAYERS WITH GOALS )

```python
balance_off = players[players['goals'] > 0].groupby('team').size().reset_index()
balance_off.columns = ['team', 'number_of_scorers']
balance_off = balance_off.sort_values('number_of_scorers', ascending=False)
print("Πιο balanced επίθεση:")
print(balance_off.head(1).to_string(index=False))  

print("\nTop 5 balanced επιθέσεις:")
print(balance_off.head(5).to_string(index=False))  
```

### FINDING WHICH IS THE TEAM WITH THE MOST CREATIVE MIDFIELDERS (TEAMS THAT HAVE MIDFIELDERS WITH MANY ASSISTS)

### THE TEAM WITH THE MOST SOLID DEFENDSE(THE LEAST YELLOW+RED CARDS)

### CORRELATION BEETWEEN MINUTES AND GOALS 


```python
correlation = players['minutes'].corr(players['goals'])
print("Correlation between minutes and goals:", correlation)
```

### CREATING A CORRELATION HEATMAP 


```python
cols_to_exclude=['born','starts','penalty_kick_attempts','goals+assists','assists_per_game_ratio','goals_per_game','minutes/played','goals/minutes']
numeric_df=players.select_dtypes(include='number')
filtered_df=numeric_df.drop(columns=cols_to_exclude)
correlation_matrix=filtered_df.corr()
fig, ax = plt.subplots(figsize=(35,10))
sns.heatmap(data=correlation_matrix,annot=True,fmt=".2f",cmap='coolwarm',square=True,lw=0.5)
plt.title('Correlation Heatmap')
# EXCLUDE SOME COLS BEACUSE THE CORRELATION HEATMAP WAS TOO DENSE 


```

# Player Of The Season

```python
players.loc[:, 'Rating'] = (
    players['goals'] * 4 + 
    players['assists'] * 3 +
    players['expected_goals'] * 2 +
    players['progressive_passes'] / 10 + 
    players['progressive_carries'] / 100 +
    players['minutes'] / 100
)

POTS=players.nlargest(1,'Rating')[['name','nation','position','team','goals','assists','expected_goals','progressive_passes','progressive_carries','minutes','Rating']]
print('Player of the season : ')
print(POTS.to_string(index=False,col_space=10))
```

### TOP 5 PLAYER OF THE SEASON NOMINEE 

```python
players.nlargest(5,'Rating')[['name','goals','assists','minutes','expected_goals','progressive_passes','progressive_carries','Rating']].reset_index(drop=True)
```

# Young Player of the Season


```python
wonderkids=players[(players['age']<=23) & (players['played']>=17)].copy()
YPOTS=wonderkids.nlargest(1 , 'Rating')[['name','nation','position','team','goals','assists','expected_goals','progressive_passes','progressive_carries','minutes','Rating']]
print('Young players of the season :')
print(YPOTS.to_string(index=False,col_space=11))
    
```
